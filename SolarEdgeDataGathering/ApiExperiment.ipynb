# Run a simple test to make sure api is working

```python
import requests
import json
import datetime
```

```python
key = ''
with open('solar_edge_api_key.txt') as file:
    key = file.read().replace('\n', '')
key
```

```python
base_url = 'https://monitoringapi.solaredge.com'
url = f'{base_url}/sites/list?api_key={key}'
response = requests.get(url)
response_dict = response.json()
print(response_dict)
```

```python
json_string = json.dumps(response_dict)
with open('site_information.json', 'w') as file:
    file.write(json_string)
```

# Get a sample api call for the energy production

```python
# for now pulling id from the saved json file, using just for testing
id = 103941
```

Get the site start and end date from the api

```python
url = f'{base_url}/site/{id}/dataPeriod?api_key={key}'
response = requests.get(url)
response_dict = response.json()
```

```python
json_string = json.dumps(response_dict)
with open('alan_knudson_start_end.json', 'w') as file:
    file.write(json_string)
```

```python
start = response_dict['dataPeriod']['startDate']
end = response_dict['dataPeriod']['endDate']
print(start)
print(end)
```

Call the energy production api

```python
url = f'{base_url}/site/{id}/energy?timeUnit=MONTH&startDate={start}&endDate={end}&api_key={key}'
response = requests.get(url)
```

```python
print(response)
response_dict = response.json()
print(response_dict)
```

Test it on 15 minute data

```python
def last_day_of_month(any_day):
    next_month = any_day.replace(day=28) + datetime.timedelta(days=4)  # this will never fail
    return next_month - datetime.timedelta(days=next_month.day)
```

```python
end = datetime.datetime.strptime(start, '%Y-%m-%d')
end = last_day_of_month(end)
print(end)
```

```python
end_string = end.strftime('%Y-%m-%d')
print(end_string)
url = f'{base_url}/site/{id}/energy?timeUnit=QUARTER_OF_AN_HOUR&startDate={start}&endDate={end_string}&api_key={key}'
```

```python
response = requests.get(url)
response_dict = response.json()
print(response_dict)
```
