## Move the combined data

```python
import shutil

for site in sites:
    site_id = site['id']
    src = f'../DataMerge/data/{site_id}/production_weather_combination.csv'
    dest = f'../../data/{site_id}/production_weather_combination.csv'
    shutil.copyfile(src, dest)
    
```

```python
import os

for site in sites:
    site_id = site['id']
    src = f'../../data/{site_id}/production_weather_combination.csv'
    dest_path = f'../../data/{site_id}/combination_data'
    if not os.path.exists(dest_path):
        os.mkdir(dest_path)
    dest = f'{dest_path}/production_weather_combination.csv'
    #shutil.copyfile(src, dest)
    
    os.rename(src, dest)
```
