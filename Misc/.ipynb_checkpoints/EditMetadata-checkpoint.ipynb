I want to create all of the ids in that folder

```python
import json

site_info = ''
with open('../SolarEdgeDataGathering/site_information.json') as file:
    site_info = file.read()
site_info = json.loads(site_info)
site_metadata = site_info['sites']['site']
site_metadata
```

```python
import os

for site in site_info['sites']['site']:
    site_id = site['id']
    folder = f'../data/{site_id}'
    os.mkdir(f'../data/{site_id}')
```

Read in the locations csv file

```python
df = pd.read_csv('../SolarEdgeDataGathering/weather-data/locations.csv')
df = df.set_index('id')
df
```

```python
for site in site_metadata:
    #row = df.loc[site['id']]
    lat = df.loc[site['id'], 'lat']
    long = df.loc[site['id'], 'long']
    site['location']['lat'] = lat
    site['location']['long'] = long
site_metadata
```

Need to get the start and end times for each system

```python
df_interval = pd.read_csv('../SolarEdgeDataGathering/site_start_end.csv')
df_interval = df_interval.set_index('id')
df_interval
```

```python
for site in site_metadata:
    start = df_interval.loc[site['id'], 'startDate']
    end = df_interval.loc[site['id'], 'endDate']
    site['time_interval'] = {
        'startDate': start, 
        'endDate': end
    }
```
